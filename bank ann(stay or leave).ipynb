```python
import numpy as np
import pandas as pd
import tensorflow as tf
```

Now after importing libraries


Data preprocessing 

```python
data=pd.read_csv('/content/Churn_Modelling.csv')

```

```python
X=data.iloc[:,3:-1].values
Y=data.iloc[:,-1].values
```

```python
print(X)
```

```python
print(Y)
```

Now as we can see in X there is categorical data and our neural network can only take numerical values
so we will encode it


```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
X[:,2]=le.fit_transform(X[:,2])
X[:,1]=le.fit_transform(X[:,1])
```

```python
print("Male as",X[771,2])
print("Female as ",X[0,2])

```

Training and testing

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(X,Y,test_size=0.2,random_state=0)
```

Feature Scaling as neural network is based on gradient descent method it will take. a large time to find the minimum of cost function

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
xtrain = sc.fit_transform(xtrain)
xtest = sc.transform(xtest)
```

Starting with the ann initialise it using tensorflow keras

```python
ann=tf.keras.models.Sequential()
```

adding layers using layers and dense method units is no of nodes we want in our hidden 1st layer

```python
ann.add(tf.keras.layers.Dense(units=6,activation='relu'))
```

adding 2nd layer

```python
#ann.add(tf.keras.layers.Dense(units=4, activation='relu'))
```

```python
#ann.add(tf.keras.layers.Dense(units=3, activation='relu'))
```

```python
ann.add(tf.keras.layers.Dense(units=6, activation='relu'))
```

adding output layer and using signoid function for the probabilty

```python
ann.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
```

```python
ann.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
ann.fit(xtrain, ytrain, batch_size = 32, epochs = 100)
```

### Predicting the Test set results

```python
y_pred = ann.predict(xtest)
y_pred = (y_pred > 0.5)
print(np.concatenate((y_pred.reshape(len(y_pred),1), ytest.reshape(len(ytest),1)),1))
```

Making confusion matrix for accuracy checking

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm=confusion_matrix(ytest,y_pred)
print(cm)
print(accuracy_score(ytest,y_pred))
```
